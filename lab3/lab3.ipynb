```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Завантаження данних

```python
tips = sns.load_dataset('tips')
tips.head(16)
```

Візуалізація даних.

Графіки для аналізу кількісних даних:
1. distplot - показує розподіл одновимірного набору спостережень.

```python
sns.distplot(tips['total_bill'])
```

2. jointplot – дозволяє співставити розподіл двох змінних та їх взаємозв’язок за допомогою діаграми розсіювання.

```python
sns.jointplot(x = "tip", y = "total_bill", data=tips, kind="hex")
```

3. pairplot – будує попарні взаємозв’язки між змінними в повному наборі даних.

```python
sns.pairplot(tips, hue = "sex")
```

Графіки для аналізу категоріальних даних:
    
1. barplot – дозволяють отримувати сукупні дані по кількісним даним в залежності від категоріальної змінної.

```python
sns.barplot(x = "sex", y = "tip", data = tips)
```

```python
tipss = tips.head(200)
sns.barplot(x = "size", y = "total_bill", data = tipss, estimator = np.median)
```

2. boxplots - використовується для візуалізації розподілу категоріальних даних за певною кількісною змінною.

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, palette="rainbow")
```

3. stripplot – представляє собою діаграму розсіювання, де одна за змінних категоріальна.

```python
sns.stripplot(x = "day", y = "total_bill", data = tips)
```

Графіки для аналізу взаємозв’язку даних:

1. Heatmap – графік взаємозв’язку (кореляції) між даними у вигляді cматриці.

```python
 sns.heatmap(tips.corr())
```

```python
g = sns.FacetGrid(tips, col = "time", row = "smoker")
g = g.map(plt.hist, "total_bill")
```
